```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
execution_times = [
    {"input_size": 0.01264, "execution_time": 22},
    {"input_size": 2.09, "execution_time": 193},
    {"input_size": 2.86, "execution_time": 310},
    {"input_size": 5.07, "execution_time": 413},
    {"input_size": 9.01, "execution_time": 585},
]

# Creare un DataFrame
df2 = pd.DataFrame(execution_times)

# Creare il grafico
fig, ax = plt.subplots(figsize=(10, 5), facecolor='white')
# Impostare lo sfondo interno del grafico come bianco
ax.set_facecolor('white')

# Personalizzare il grafico
ax.plot(df2['input_size'], df2['execution_time'], marker='o', linestyle='-', color='lightskyblue', markersize=8, linewidth=2)

# Aggiungere il bordo nero
for spine in ax.spines.values():
    spine.set_edgecolor('black')
    spine.set_linewidth(1.5)

# Etichette degli assi
ax.set_xlabel("Input size (GB)", fontsize=12, labelpad=15)
ax.set_ylabel("Execution time (sec)", fontsize=12, labelpad=15)

# Titolo del grafico
ax.set_title("Execution time of data profiling on AWS as a function of input size", fontsize=16, pad=20)

# Aggiungere una griglia sottile
ax.grid(True, linestyle='--', linewidth=0.5, color='lightgray')

plt.show()
```
