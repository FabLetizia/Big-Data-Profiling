```python
import matplotlib.pyplot as plt
import pandas as pd
```

# Dupicate rows

```python
datasets = [
    "airline.csv", "airline_cleaned.csv",
    "titles.csv", "titles_cleaned.csv",
    "books_rating.csv", "books_rating_cleaned.csv",
    "london.csv", "london_cleaned.csv",
    "2019-Nov.csv", "2019-Nov_cleaned.csv"
]

duplicate_percentages = [
    0, 0,  # airline.csv, airline_cleaned.csv
    16.7, 0,  # titles.csv, titles_cleaned.csv
    0.3, 0,  # books_rating.csv, books_rating_cleaned.csv
    0, 0,  # london.csv, london_cleaned.csv
    0.1, 0  # 2019-Nov.csv, 2019-Nov_cleaned.csv
]

x_labels = [dataset.replace('.csv', '') for dataset in datasets]
x = range(len(x_labels))
y = duplicate_percentages

colors = ['skyblue' if 'cleaned' not in dataset else 'lightgreen' for dataset in datasets]

plt.figure(figsize=(12, 10))
bars = plt.bar(x, y, color=colors)

plt.xlabel('Datasets')
plt.ylabel('Duplicate Percentages')
plt.title('Duplicate Percentages in Datasets')
plt.xticks(x, x_labels, rotation=45, ha='right') 

for i, bar in enumerate(bars):
    yval = y[i]
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.5, f'{yval}%', ha='center', va='bottom')

plt.legend(handles=[bars[0], bars[1]], labels=['Original', 'Cleaned'])
plt.tight_layout()
plt.show()
```

# Missing cells

```python
datasets = [
    "airline.csv", "airline_cleaned.csv",
    "titles.csv", "titles_cleaned.csv",
    "books_rating.csv", "books_rating_cleaned.csv",
    "london.csv", "london_cleaned.csv",
    "2019-Nov.csv", "2019-Nov_cleaned.csv"
]

null_percentages = [
    0, 0,  # airline.csv, airline_cleaned.csv
    66.7, 0.1,  # titles.csv, titles_cleaned.csv
    12.1, 12.1,  # books_rating.csv, books_rating_cleaned.csv
    0.1, 0.1,  # london.csv, london_cleaned.csv
    5.2, 5.1  # 2019-Nov.csv, 2019-Nov_cleaned.csv
]

x_labels = [dataset.replace('.csv', '') for dataset in datasets]
x = range(len(x_labels))
y = null_percentages

colors = ['skyblue' if 'cleaned' not in dataset else 'lightgreen' for dataset in datasets]

plt.figure(figsize=(12, 6))
bars = plt.bar(x, y, color=colors)

plt.xlabel('Datasets')
plt.ylabel('Null Percentages')
plt.title('Null Percentages in Datasets')
plt.xticks(x, x_labels, rotation=45, ha='right')  

for i, bar in enumerate(bars):
    yval = y[i]
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.5, f'{yval}%', ha='center', va='bottom')

plt.legend(handles=[bars[0], bars[1]], labels=['Original', 'Cleaned'])
plt.tight_layout()
plt.show()
```

# Execution times

### Initial Datasets 

```python
# Data for initial execution times
execution_times = [
    {"input_size": 0.01264, "execution_time": 12.94},
    {"input_size": 2.09, "execution_time": 208.42},
    {"input_size": 2.86, "execution_time": 344.94},
    {"input_size": 5.07, "execution_time": 417.23},
    {"input_size": 9.01, "execution_time": 554.35},
]

# initial_datasets labels
initial_datasets = [
    "airline.csv",
    "titles.csv",
    "books_rating.csv",
    "london.csv",
    "2019-Nov.csv"
]

df_initial = pd.DataFrame(execution_times)
fig, ax = plt.subplots(figsize=(10, 5), facecolor='white')
ax.set_facecolor('white')

# Plot the data
ax.plot(df_initial['input_size'], df_initial['execution_time'], marker='o', linestyle='-', color='lightseagreen', markersize=8, linewidth=2)

for i, txt in enumerate(initial_datasets):
    if txt == "2019-Nov.csv":
        ax.annotate(txt, (df_initial['input_size'][i], df_initial['execution_time'][i]), textcoords="offset points", xytext=(15, -20), ha='right', fontsize=10)
    elif txt == "airline.csv":
        ax.annotate(txt, (df_initial['input_size'][i], df_initial['execution_time'][i]), textcoords="offset points", xytext=(60, -2), ha='right', fontsize=10)
    else:
        ax.annotate(txt, (df_initial['input_size'][i], df_initial['execution_time'][i]), textcoords="offset points", xytext=(5, -15), ha='left', fontsize=10)

for spine in ax.spines.values():
    spine.set_edgecolor('black')
    spine.set_linewidth(1.5)

# Set labels and title
ax.set_xlabel("Input size (GB)", fontsize=12, labelpad=15)
ax.set_ylabel("Execution time (sec)", fontsize=12, labelpad=15)
ax.set_title("Execution time of (initial) data profiling on AWS as a function of input size", fontsize=16, pad=20)

ax.grid(True, linestyle='--', linewidth=0.5, color='lightgray')

plt.tight_layout()
plt.show()
```

### Cleaned Datasets

```python
execution_times_cleaned = [
    {"input_size": 0.01264, "execution_time": 12.94},
    {"input_size": 2.09, "execution_time": 127.06},
    {"input_size": 2.86, "execution_time": 277.12},
    {"input_size": 5.07, "execution_time": 388.59},
    {"input_size": 9.01, "execution_time": 439.88},
]

cleaned_datasets = [
    "airline_cleaned.csv",
    "titles_cleaned.csv",
    "books_rating_cleaned.csv",
    "london_cleaned.csv",
    "2019-Nov_cleaned.csv"
]

df_cleaned = pd.DataFrame(execution_times_cleaned)
fig, ax = plt.subplots(figsize=(10, 5), facecolor='white')
ax.set_facecolor('white')
ax.plot(df_cleaned['input_size'], df_cleaned['execution_time'], marker='o', linestyle='-', color='mediumvioletred', markersize=8, linewidth=2)

for i, txt in enumerate(cleaned_datasets):
    if txt == "2019-Nov_cleaned.csv":
        ax.annotate(txt, (df_cleaned['input_size'][i], df_cleaned['execution_time'][i]), textcoords="offset points", xytext=(20, -25), ha='right', fontsize=10)
    elif txt == "airline_cleaned.csv":
        ax.annotate(txt, (df_cleaned['input_size'][i], df_cleaned['execution_time'][i]), textcoords="offset points", xytext=(110, -2), ha='right', fontsize=10)
    else:
        ax.annotate(txt, (df_cleaned['input_size'][i], df_cleaned['execution_time'][i]), textcoords="offset points", xytext=(5, -15), ha='left', fontsize=10)
    
for spine in ax.spines.values():
    spine.set_edgecolor('black')
    spine.set_linewidth(1.5)

ax.set_xlabel("Input size (GB)", fontsize=12, labelpad=15)
ax.set_ylabel("Execution time (sec)", fontsize=12, labelpad=15)
ax.set_title("Execution time of (cleaned) data profiling on AWS as a function of input size", fontsize=16, pad=20)

ax.grid(True, linestyle='--', linewidth=0.5, color='lightgray')
plt.show()
```
